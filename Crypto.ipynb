# Manipuler les données 

```python
import seaborn as sns
```

```python
# spécifier les parametres 
currency_n = 'EUR'
metric = 'Close'
crypto_n = ['BTC','ETH','LTC','XRP','DASH','SC','DOGE']
colnames = []
```

```python
first = True
for ticker in crypto_n:
    data = web.DataReader(f"{ticker}-{currency_n}","yahoo",start,end)
    if first : 
        combined = data[[metric]].copy()
        colnames.append(ticker)
        combined.columns = colnames
        first = False
    else : 
        combined = combined.join(data[metric])
        colnames.append(ticker)
        combined.columns = colnames 
combined
```

```python
combined.to_csv('crypto_v1.csv')
```

```python
for ticker in crypto_n : 
    plt.plot(combined[ticker], label=ticker)
    plt.legend(loc = 'upper right')
    plt.show()
```

```python
#Visualisation des crypto monnaies sur une echelle logarithmique
plt.yscale('log')
for ticker in crypto_n : 
    plt.plot(combined[ticker], label=ticker)
plt.legend(loc = 'upper left')
plt.show()
```

```python
# correlation 
combined = combined.pct_change().corr(method='pearson')
sns.heatmap(combined, annot=True, cmap='coolwarm')
plt.show()
```
